```python
def sim_overtime():
    #Setting variables
    time_elapsed=0 #update this each drive
    possessions = np.array([])
    drive_outcomes = np.array(["TD", "FG", "0"])
    curr_team = 1
    
    #Simulating first drive and adding it to drive_outcomes:
    curr_drive = np.random.choice(drive_outcomes, p=[.21,.13,.66])
    possessions = np.append(possessions, curr_drive)
    time_elapsed += stats.expon.rvs(scale=2.5)
    
    #Team 1 scores a touchdown
    if curr_drive == "TD":
        return 1, possessions
    #Team 1 scores a field goal
    elif curr_drive == "FG":
        curr_team = 2
        curr_drive = np.random.choice(drive_outcomes, p=[.21,.13,.66])
        possessions = np.append(possessions, curr_drive)
        time_elapsed += stats.expon.rvs(scale=2.5)
        #Team 2 scores a touchdown
        if curr_drive == "TD":
            return 2, possessions
        elif curr_drive == "0":
            return 1, possessions 
        
    #Either team 1 does not score or both teams score field goals:
    while time_elapsed < 10:
        #Setting the current team
        if curr_team == 1:
            curr_team = 2
        else:
            curr_team = 1
        #Calculating remaining drive components
        curr_drive = np.random.choice(drive_outcomes, p=[.21,.13,.66])
        possessions = np.append(possessions, curr_drive)
        time_elapsed += stats.expon.rvs(scale=2.5)
        #Current team scores
        if curr_drive != "0":
            return curr_team, possessions
        
    #If the loop breaks, the time runs out and the overtime results in a tie
    return 0, possessions
```

```python
#Initializing arrays
sim_winners = np.array([])
sim_possessions = np.array([])
sim_num = 0

#Performing simulations
while sim_num < 10000:
    curr_sim = sim_overtime()
    sim_winners = np.append(sim_winners, curr_sim[0])
    sim_possessions = np.append(sim_possessions, len(curr_sim[1]))
    sim_num += 1
    
#Reporting probabilities
prob_1 = np.sum(sim_winners == 1) / len(sim_winners)
prob_2 = np.sum(sim_winners == 2) / len(sim_winners)
prob_0 = np.sum(sim_winners == 0) / len(sim_winners)

print("The probability of Team 1 winning is approximately", prob_1)
print("The probability of Team 2 winning is approximately", prob_2)
print("The probability of the teams drawing is approximately", prob_0)

print(np.sum(sim_winners == 1))
print(np.sum(sim_winners == 2))
```
