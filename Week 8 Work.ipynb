```python
import numpy as np 
import pandas as pd 
from scipy import stats
import matplotlib.pylab as plt 
%matplotlib inline
```

```python
#Loading the file into a DataFrame
file_path = 'income_data.csv'
dfIncomeData = pd.read_csv(file_path)
dfIncome = dfIncomeData.loc[dfIncomeData.index[:5000], ["Income"]]

#Plotting a histogram
fig, ax= plt.subplots()
ax.hist(dfIncome, color="steelblue", edgecolor="white", bins=20)
```

```python
print("The mean income considering every member of the population is:")
print(dfIncomeData["Income"].mean())
```

```python
def income_sample_mean(df, n):
    dfSample = df.sample(n)
    return dfSample["Income"].mean()
```

```python
#your code here
print("With a sample size of 10, the function generates the following 5 mean approximations:")
for x in range(5):
    print(income_sample_mean(dfIncome, 10))
print("")
print("Running the same function with a sample size of 1,000 yeilds the following 5 mean approximations:")
for x in range(5):
    print(income_sample_mean(dfIncome, 1000))
    
```

```python
def income_sample_dist(df, n, num_means):
    #Your code here
    data = np.zeros(1000)
    for x in range(num_means):
        data[x - 1] = income_sample_mean(df, n)
    data = pd.DataFrame(data)
    #Include code to draw a histogram of results
    fig, ax = plt.subplots(figsize=(5,3))
    ax.hist(data, bins=20)
    ax.set_title(n)
```

```python
#Now run the code for sample sizes of n=5, n=50 and n=500 and compare the histograms
income_sample_dist(dfIncome, 5, 1000)
income_sample_dist(dfIncome, 50, 1000)
income_sample_dist(dfIncome, 500, 1000)
```
