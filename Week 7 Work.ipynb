```python
import numpy as np
import math
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#Calculating the critical value
print("SOLUTION:")
print("x = P(Z < 0.95) implies x =", stats.norm.ppf(0.95), ", meaning the area under the curve to the right of this x value = 0.05")
```

```python
#Plotting Graph
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x,0,1))

#Adding shaded area under graph
plt.fill_between(x, stats.norm.pdf(x,0,1), where=[(x > 1.645) for x in x], label='area = 0.05', alpha=0.4)

#Adding legend
plt.legend(loc=(.70,.6), shadow=True)

#Adding labels
labels=[-4, 1.645, 4]
my_xticks=[-4,"critical value = appx 1.645",4]
plt.xticks(labels,my_xticks)
```
